```python
import pandas as pd
import numpy as np

print("=" * 80)
print("DATA TYPE ISSUE DETECTED")
print("=" * 80)

# Check data types
print(f"Embedding sample_id dtype: {emdata['sample_id'].dtype}")
print(f"Metrics sample_id dtype: {metdata['sample_id'].dtype}")

print(f"\nExample values:")
print(f"  Embedding: {emdata['sample_id'].iloc[0]} (type: {type(emdata['sample_id'].iloc[0])})")
print(f"  Metrics:   {metdata['sample_id'].iloc[0]} (type: {type(metdata['sample_id'].iloc[0])})")

# FIX: Convert both to the same type (integers)
emdata['sample_id'] = emdata['sample_id'].astype(int)
metdata['sample_id'] = metdata['sample_id'].astype(int)

print("\n" + "=" * 80)
print("AFTER TYPE CONVERSION")
print("=" * 80)

# Now sort and compare
emdata_sorted = emdata.sort_values('sample_id').reset_index(drop=True)
metdata_sorted = metdata.sort_values('sample_id').reset_index(drop=True)

emdata_ids = emdata_sorted['sample_id']
metdata_ids = metdata_sorted['sample_id']

print(f"\n1. Are they identical after sorting and type conversion?")
are_equal = emdata_ids.equals(metdata_ids)
print(f"   {are_equal}")

print(f"\n2. Length comparison:")
print(f"   Embedding IDs: {len(emdata_ids)}")
print(f"   Metrics IDs: {len(metdata_ids)}")

print(f"\n3. First 10 sorted IDs (now both integers):")
print(f"   Embedding: {list(emdata_ids.head(10))}")
print(f"   Metrics:   {list(metdata_ids.head(10))}")

# Check for differences
em_set = set(emdata_ids)
met_set = set(metdata_ids)

print(f"\n4. Set comparison:")
print(f"   Unique IDs in embedding: {len(em_set)}")
print(f"   Unique IDs in metrics: {len(met_set)}")
print(f"   Common IDs: {len(em_set & met_set)}")

only_in_em = em_set - met_set
only_in_met = met_set - em_set

print(f"   IDs only in embedding: {len(only_in_em)}")
if only_in_em:
    print(f"   Examples: {sorted(list(only_in_em))[:10]}")

print(f"   IDs only in metrics: {len(only_in_met)}")
if only_in_met:
    print(f"   Examples: {sorted(list(only_in_met))[:10]}")

print("\n" + "=" * 80)
print("RECOMMENDATION")
print("=" * 80)

if are_equal:
    print("✓ Perfect match! Use sorted datasets directly.")
else:
    common_count = len(em_set & met_set)
    print(f"⚠ Only {common_count} common samples found.")
    print(f"  Lost {len(emdata) - common_count} from embedding ({(len(emdata) - common_count)/len(emdata)*100:.1f}%)")
    print(f"  Lost {len(metdata) - common_count} from metrics ({(len(metdata) - common_count)/len(metdata)*100:.1f}%)")
    print("\n  Use inner merge to align datasets:")
    print("  merged = emdata.merge(metdata, on='sample_id', how='inner')")

print("=" * 80)
```
